# Stock Price Analysis using Monte Carlo Simulations

The aim of this notebook is to explore how to use Monte Carlo simulations in order to predict future prices of stocks and also get a sense of the likelihood of those predictions.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')

import math
from pandas_datareader import data
```

## Apple (AAPL)

<img src="apple_logo.jpg">

First we will start with a well known company and also a personal holding of mine in Apple.

```python
# get historical data from last 10 years (this can be adjusted)
apple = data.DataReader('AAPL', 'yahoo', start='1/1/2009')
apple.head()
```

```python
# check columns are correct data types
apple.info()
```

```python
plt.figure(figsize=(12,5))
plt.plot(apple['Adj Close'])
plt.title('AAPL Historical Prices')
plt.show()
```

### Collect Annual Mean and Standard Dev

Now we will need to calculate the mean (CAGR) and the standard deviation (Annualized Volatility) which will be used to feed into the Monte Carlo simulation.

```python
# calculate time elapsed
time_elapsed = (apple.index[-1] - apple.index[0]).days
print('Time elapsed: {} days'.format(time_elapsed))

number_of_years = time_elapsed/365
print('Number of years:', np.round(number_of_years, 1))

# calculate TWR (Time Weighted Return)
twr = (apple['Adj Close'][-1] / apple['Adj Close'][0])
print('Time weighted return: {:.2%}'.format(twr))

# calculate compound annual growth rate to be our mean return
cagr = twr**(1/number_of_years) - 1
print('Compound annual growth rate: {:.2%}'.format(cagr))
```

```python
# calculate standard deviation of daily price changes
std_dev = apple['Adj Close'].pct_change().std()

# calculate standard deviation of annual price changes
# Roughly 252 trading days in a year
# reference: https://www.fool.com/knowledge-center/how-to-calculate-annualized-volatility.aspx
number_of_trading_days = 252
std_dev = std_dev*math.sqrt(number_of_trading_days)
print('Standard deviation of annual price change: {:.2%}'.format(std_dev))
```

### Random Walk

Now that we have our annual mean and standard deviation, we can perform what is known as a random walk. This basically uses a random normal distribution of values based on the mean and standard deviation provided in order to produce a projected price trajectory over the next trading year (252 days).

```python
# plant seed for reproducibility
np.random.seed(42)

# using the annual mean and standard deviation
# generate an array of random values over the next trading year (252 days)
daily_return_percentages = np.random.normal(cagr/number_of_trading_days, std_dev/math.sqrt(number_of_trading_days), 
                                            number_of_trading_days)+1

# with this array of random future daily returns, can add this to the last actual closing price
# this generates what is known as a random walk
price_series = [apple['Adj Close'][-1]]

for i in daily_return_percentages:
    price_series.append(price_series[-1] * i)

# plot results of the random walk
plt.figure(figsize=(12,5))
plt.plot(price_series)
plt.title('AAPL Random Walk')
plt.ylabel('Price')
plt.xlabel('Days')
plt.show()
```

This chart looks pretty good, but remember this is only one random walk so we shouldn't get our hopes up and rush out to buy Apple stock right this second.  Our analysis will be better if we run this simulation many times over in order to get a more accurate overall outcome.

### Monte Carlo Simulation

```python
np.random.seed(42)

# number of random walks
number_of_trials = 3000

# make additonal list of closing prices to get a distribution of all possible outcomes
closing_prices = []

# produce random walks
for i in range(number_of_trials):
    daily_return_percentages = np.random.normal(cagr/number_of_trading_days, 
                                                std_dev/math.sqrt(number_of_trading_days), number_of_trading_days)+1
    price_series = [apple['Adj Close'][-1]]
    
    for j in daily_return_percentages:
        price_series.append(price_series[-1] * j)
    
    # add each random walk closing price
    closing_prices.append(price_series[-1])
    
    # plot random walks
    plt.plot(price_series)

plt.show()

# plot closing prices distribution
plt.hist(closing_prices, bins=40)
plt.show()
```

Above we can see all of the random walks our Monte Carlo simulation produced along with the each of their year end closing prices as a histogram.

### Analyze Results

```python
# calculate average closing price
mean_end_price = np.round(np.mean(closing_prices), 2)
print('Average closing price:', mean_end_price)

# calculate 90th and 10th percentile prices
top_ten = np.percentile(closing_prices, 100-10)
bottom_ten = np.percentile(closing_prices, 10)

plt.figure(figsize=(12,5))
plt.hist(closing_prices, bins=40, alpha=0.9, color='silver')
plt.axvline(top_ten, color='red', linestyle='--', linewidth=2, label='90th %')
plt.axvline(bottom_ten, color='red', linestyle='--', linewidth=2, label='10th %')
plt.axvline(apple['Adj Close'][-1], color='green', linestyle='--', linewidth=2, label='current price')

plt.title('AAPL Distribution of Closing Prices', weight='bold', size=14)
plt.xlabel('Closing Price')
plt.legend()
plt.show()
```

Now after running the simulation 3000 times, I can be more confident in the projected bullish signal of Apple. Based on our results, the average closing price over the next year is expected to be 290.86 meaning there is a 50% chance the price will end up above or below that price target. Also looking at the chart above, we can be 90% confident that the price will finish above 200 which is very close to where the price currently is as of 8/22/2019. Being an Apple owner this gives me confidence to keep holding for the long term.

## Frontdoor Inc. (FTDR)

<img src="frontdoor_logo.jpg">

This company recently became public almost a year ago, so we will have a lot less historical data to work with.

```python
frontdoor = data.DataReader('ftdr', 'yahoo')
frontdoor.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(frontdoor['Adj Close'], color='#FF5A00')
plt.title('FTDR Historical Prices')
plt.show()
```

We can see with being a newer stock there seems to be a lot more volatility so it will be interesting to see how a Monte Carlo simulation performs.

```python
# calculate time elapsed
time_elapsed = (frontdoor.index[-1] - frontdoor.index[0]).days
print('Time elapsed: {} days'.format(time_elapsed))

number_of_years = time_elapsed/365
print('Number of years:', np.round(number_of_years, 1))

# calculate TWR (Time Weighted Return)
twr = (frontdoor['Adj Close'][-1] / frontdoor['Adj Close'][0])
print('Time weighted return: {:.2%}'.format(twr))

# calculate compound annual growth rate to be our mean return
cagr = twr**(1/number_of_years) - 1
print('Compound annual growth rate: {:.2%}'.format(cagr))
```

Note we will have to change the number of trading days variable since Frontdoor hasn't been public for over a year.

```python
# calculate standard deviation of daily price changes
std_dev = frontdoor['Adj Close'].pct_change().std()

# calculate standard deviation of annual price changes
# Hasn't been a full year so will use the amount of days been trading (237)
number_of_trading_days = len(frontdoor)
std_dev = std_dev*math.sqrt(number_of_trading_days)
print('Standard deviation of annual price change: {:.2%}'.format(std_dev))
```

```python
np.random.seed(42)

# number of random walks
number_of_trials = 3000

# make additonal list of closing prices to get a distribution of all possible outcomes
closing_prices = []

# produce random walks
for i in range(number_of_trials):
    daily_return_percentages = np.random.normal(cagr/number_of_trading_days, 
                                                std_dev/math.sqrt(number_of_trading_days), number_of_trading_days)+1
    price_series = [frontdoor['Adj Close'][-1]]
    
    for j in daily_return_percentages:
        price_series.append(price_series[-1] * j)
    
    # add each random walk closing price
    closing_prices.append(price_series[-1])
    
    # plot random walks
    plt.plot(price_series)

plt.show()

# plot closing prices distribution
plt.hist(closing_prices, bins=40)
plt.show()
```

```python
# calculate average closing price
mean_end_price = np.round(np.mean(closing_prices), 2)
print('Average closing price:', mean_end_price)

# calculate 90th and 10th percentile prices
top_ten = np.percentile(closing_prices, 100-10)
bottom_ten = np.percentile(closing_prices, 10)

plt.figure(figsize=(12,5))
plt.hist(closing_prices, bins=40, alpha=0.9, color='#FF5A00')
plt.axvline(top_ten, color='grey', linestyle='--', linewidth=2, label='90th %')
plt.axvline(bottom_ten, color='grey', linestyle='--', linewidth=2, label='10th %')
plt.axvline(frontdoor['Adj Close'][-1], color='green', linestyle='--', linewidth=2, label='current price')

plt.title('FTDR Distribution of Closing Prices', weight='bold', size=14)
plt.xlabel('Closing Price')
plt.legend()
plt.show()
```

Another bullish looking stock with an average closing price of 112.98 and also with the current price right above the 10th percentile price. Might take this as a good sign to do further research into the company in order to see if I'll buy this stock.

## Conclusion

Although these findings look extremely bullish, remember that there are many other factors that go into the pricing of a stock rather than just historical price trends. So you shouldn't base your stock picks solely on Monte Carlo simulations. Intead use these Monte Carlo distribution charts in order to gain a better understanding of the range of possible outcomes and then along with your own risk profile you can base your decisions that way.
